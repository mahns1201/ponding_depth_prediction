```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
learning_rate = 0.01
trainning_steps = 10000000
display_step = 10000
```

```python
data = np.loadtxt("parsed_alldata_np.csv", delimiter = ',', dtype = np.float32)

X = data[:10]

Y = data[-1]
```

```python
w = tf.Variable(tf.random.uniform([1, len(X)], 0, 1))

b = tf.Variable(random.random())
```

```python
def compute_loss():
    y_pred = tf.matmul(w, X) + b
    loss = tf.reduce_mean((Y - y_pred) ** 2)
    return loss
```

```python
optimizer = tf.optimizers.Adam(lr = learning_rate)
```

```python
for step in range(1, trainning_steps + 1):
    optimizer.minimize(compute_loss, var_list = [w, b])
    
    if step % display_step == 0:
        print(step, 'w:', w.numpy(), 'b:', b.numpy(), 'loss:', compute_loss().numpy())
```

```python
def rmse(loss):
    rmse = np.sqrt(loss)
    
    return rmse
```
