```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
pd_data = pd.read_csv("pd_wl3_2019.csv")
```

```python
print(pd_data)
```

```python
# 데이터의 갯수 = 2148개

x = pd_data['WL3(mm)']   
y = pd_data['PD'] 
```

```python
plt.figure(figsize = (15, 5))
plt.plot(x, y, 'bo')
plt.title('WL1 - PD Relation')
plt.xlabel('WL1')
plt.ylabel('PD')
plt.show()
```

```python
learning_rate = 0.01
trainning_steps = 30000
display_step = 1000
```

```python
data = np.loadtxt("pd_wl1_2019_np.csv", delimiter = ',', dtype = np.float64)

X = data[:, [0]]
Y = data[:, [-1]]
```

```python
a = tf.Variable(random.random())
b = tf.Variable(random.random())
```

```python
def compute_loss():
    y_pred = a * X + b
    loss = tf.reduce_mean((Y - y_pred) ** 2)
    return loss
```

```python
optimizer = tf.optimizers.Adam(lr = learning_rate)
```

```python
for step in range(1, trainning_steps + 1):
    optimizer.minimize(compute_loss, var_list = [a, b])
    
    if step % display_step == 0:
        print(step, 'a:', a.numpy(), 'b:', b.numpy(), 'loss:', compute_loss().numpy())
```

```python
line_x = np.arange(min(X), max(X), 0.01)
line_y = a * line_x + b
```

```python
plt.figure(figsize = (15, 5))
plt.plot(line_x, line_y, 'r-', label = 'predict_line')
plt.plot(X, Y, 'bo', label = 'wl3-pd relation')
plt.xlabel('WL3')
plt.ylabel('PD')
plt.legend(loc = 'upper left')
plt.show()
```
